```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dfs=pd.read_csv("sales_data.csv",encoding='iso-8859-1')dfs
```

fetching null data from amount col

```python
dfs['Amount'].isnull() # to check the null values
```

rename the column material_status to marital

```python
dfs.rename(columns={'Marital_Status':'Marital'},inplace=True)
dfs
```

change the data in the particular column

```python
dr={0:'single',1:'married'}
dfs['Marital']=dfs['Marital'].map(dr)
dfs
```

changing a particular value

```python
dr={'F':'Female','M':'Male'}
dfs['Gender']=dfs['Gender'].map(dr)
dfs
```

droping unwanted columns such as status and unnamesd1

```python
dfs.drop(['Status','unnamed1'],axis=1,inplace=True)
dfs
```

deleting the all null values in the data frame

```python
dfs.dropna(inplace=True)
dfs
```

visualize the count of males and females based on gender column using serborn

```python
plt.figure(figsize=(5,5))
b=sns.histplot(dfs['Gender'],color='lightpink',edgecolor='black',label='cube')
for val in b.containers:
  b.bar_label(val)
plt.legend(facecolor='grey')
plt.title('NUMBER OF MALES AND NUMBER OF FEMALES',color='blue')
plt.show()
```

**CONCLUSION:** In above graph to represent the how many males people and how many females peopels present in data frame

```python
gb=dfs.groupby(['Gender','Age Group','State','Occupation','Product_Category','Marital'])['Amount'].sum()
gb
```

group data for analysis

```python
g=dfs.groupby('Gender')['Amount'].sum()
gg=g.round(2)
ggg=sns.barplot(x=gg.index,y=gg.values,color='lightpink',edgecolor='green',label='gender speding by amount',palette='coolwarm')
for val in ggg.containers:
  ggg.bar_label(val)
plt.legend(facecolor='grey')
plt.title('Total speding by gender',color='green')
plt.show()
ggg
g
```

**Gender:** In above graph represent the females peoeples spending more amount compared to the males peoples.

```python
plt.figure(figsize=(10,10))
g=dfs.groupby('Age Group')['Amount'].sum()
ga=g.round(1)
gga=sns.barplot(x=ga.index,y=ga.values,color='lightpink',edgecolor='grey',palette='magma')
for val in gga.containers:
  gga.bar_label(val)
plt.legend(facecolor='grey')
plt.title('Total speding by Age_group',color='green')
plt.show()
```



**CONCLUSION:** In above graph represent the which age range peoples spending more amount.26-35 age group persond mores spending amout.





```python
plt.figure(figsize=(15,15))
gss=dfs.groupby('State')['Amount'].sum()
gs=gss
ggs=sns.barplot(x=gs.index,y=gs.values,color='lightpink',edgecolor='blue',palette='magma')
for val in ggs.containers:
  ggs.bar_label(val)
plt.legend(facecolor='grey')
plt.title('Total speding by state',color='green')
plt.xticks(rotation=90)
plt.show()
```

**CONCLUSION:** In above graph represent the which ststae mores pending amount in utter pradesh more spending th amount.

```python
plt.figure(figsize=(15,15))
go=dfs.groupby('Occupation')['Amount'].sum()
go
ggo=sns.barplot(x=go.index,y=go.values,color='lightpink',edgecolor='blue',palette='magma')
for val in ggo.containers:
  ggo.bar_label(val)
plt.legend(facecolor='grey')
plt.xticks(rotation=90)
plt.title('Total speding by Occupation',color='green')
plt.show()
```



**CONCLUSION :** Determines which profession has the highest spending behavior.In above graph represent the IT-sector more spending amount



```python
plt.figure(figsize=(20,20))
gp=dfs.groupby('Product_Category')['Amount'].sum()
gp
ggp=sns.barplot(x=gp.index,y=gp.values,color='lightpink',edgecolor='blue',palette='rainbow')
for val in ggp.containers:
  ggp.bar_label(val)
plt.legend(facecolor='grey')
plt.xticks(rotation=90)
plt.title('Total speding by Product_Category',color='red')
plt.show()
```

**CONCLUSION:** In above graph represent the how much investiment the particular product.in above graph more investiment the food product
